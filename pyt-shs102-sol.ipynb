## Partie 1 — Boucles et conditions

### 1.1 La boucle `for`

Les boucles permettent d'effectuer une même séquence d'opérations sur différents éléments avec quelques lignes de code seulement. C'est l'un des mécanismes qui rend la programmation réellement utile pour traiter des données en recherche.

#### Structure d'une boucle

<p align="center">
  <img src="fig/boucle.png" alt="Photo centrée" width="600">
</p>

- **`variable (i)`** : le nom donné à l'élément courant à chaque tour de boucle. Ce nom est arbitraire (`i`, `ligne`, `participant`, etc.).
- **`iterable`** : tout objet que Python peut parcourir élément par élément :
  - Les listes : `['F', 'M', 'F']`
  - Les chaînes de caractères : `"Employé"`
  - Les dictionnaires : `{"clé": "valeur"}`
  - Les ensembles (*sets*) : `{25, 33, 47}`
  - Les objets retournés par `range()`
  - Les fichiers ouverts
- La ligne se termine par un **`:`**, et le corps de la boucle est **indenté** (décalé vers la droite). Dans un notebook, l'indentation se fait automatiquement après le `:`.
- Python détecte automatiquement la fin de l'itérable et quitte la boucle.

#### Exemple — Normalisation d'un titre de recherche


````python
# Un extrait de texte brut provenant d'un rapport de recherche
titre_brut = "données sociales. politiques publiques. accès au logement"

# Afficher chaque caractère en majuscules
for caractere in titre_brut:
    print(caractere.upper())
````

### 1.2 La déclaration conditionnelle `if`

La déclaration `if` permet d'exécuter un bloc de code **seulement si une condition est vraie**. Elle introduit de la logique décisionnelle dans un programme.

#### Syntaxe de base

```python
if condition:
    # bloc exécuté si la condition est True
```

Comme pour les boucles, le code suivant le `if` doit être indenté.

#### Opérateurs disponibles

| Opérateur | Signification |
|-----------|---------------|
| `==` | égal à |
| `!=` | différent de |
| `<`, `>`, `<=`, `>=` | inférieur / supérieur |
| `and`, `or`, `not` | opérateurs logiques |
| `in` | vérifie si un élément existe dans une séquence |

#### Exemples


````python
# Vérification du critère de majorité civile
age_repondant = 18

if age_repondant >= 18:
    print("Éligible pour participer à l'étude électorale.")
````

````python
# Si la condition est False, le bloc ne s'exécute pas
taille_echantillon = 11

# Vérifier si le groupe respecte la limite méthodologique
if taille_echantillon <= 10:
    print("Le groupe est considéré comme un petit échantillon qualitatif.")
````

````python
# Vérifier si un jour appartient à une liste
jours_repos = ['samedi', 'dimanche']

if 'lundi' in jours_repos:
    print("Cette journée fait partie du temps de repos.")
````

#### La clause `else`

On peut ajouter un bloc `else` pour exécuter du code quand la condition `if` est `False`.


````python
jours_repos = ['samedi', 'dimanche']
jour_observe = 'lundi'

# Classification automatique de la donnée
if jour_observe in jours_repos:
    print("Cette journée fait partie du temps de repos.")
else:
    print("Cette journée fait partie de la semaine de travail.")
````

### 2.2 Fichiers texte au format non tabulaire

Certains fichiers ne sont pas tabulaires ou ont des lignes non uniformes (journaux de terrain, transcriptions, données biologiques `.txt`, `.fasta`, etc.). On les lit avec la fonction `open()`.

#### Structure générale

```python
with open('chemin/vers/fichier.txt', 'r') as f:
    for line in f:
        line = line.strip()
        # traiter la ligne
```

**Décortiquons chaque composant :**

| Élément | Rôle |
|---------|------|
| `with open(...)` | Ouvre le fichier et garantit qu'il sera fermé même en cas d'erreur |
| `'r'` | Mode lecture (*read*). Autres modes : `'w'` (écriture), `'a'` (ajout) |
| `as f` | Donne un nom à l'objet fichier (le nom `f` est arbitraire) |
| `for line in f` | Parcourt le fichier ligne par ligne |
| `.strip()` | Supprime les espaces et sauts de ligne `\n` en début et fin de ligne |
| `.append(line)` | Ajoute la ligne nettoyée à une liste |

#### Exemple — Lire un fichier de catégories


````python
categories = []

with open('data/categories.txt', 'r') as f:
    for line in f:
        line = line.strip()
        categories.append(line)
````

````python
# Afficher le contenu de la liste
categories
````

### 🖊️ Exercice 2.1 — Filtrer des participants par âge

Lisez le fichier `participants.txt` (format : `Nom,Age` par ligne), extrayez l'âge et affichez uniquement les participants de **plus de 30 ans**.

> 💡 **Indice :** `.strip()` nettoie les sauts de ligne, `.split(',')` divise la ligne en deux éléments (`donnees[0]` = nom, `donnees[1]` = âge), et `int()` convertit l'âge en nombre pour pouvoir le comparer.

````python
with open('data/participants.txt', 'r') as f:
    for line in f:
        # Nettoyer et séparer le nom de l'âge
        donnees = line.strip().split(',')
        nom_participant = donnees[0]
        age = int(donnees[1])

        # Filtrer les participants de plus de 30 ans
        if age > 30:
            print(nom_participant, age)
````

### 🖊️ Exercice 2.2 — Calculer une moyenne à partir d'un fichier

Lisez le fichier `travail_domestique.txt` (un nombre entier par ligne, en minutes) et calculez la **moyenne du temps de travail domestique quotidien**.

> 💡 **Indice :** Les fonctions `sum()` et `len()` appliquées à une liste permettent de calculer une moyenne.


````python
temps_domestique = []

with open('data/travail_domestique.txt', 'r') as f:
    for ligne in f:
        ligne = ligne.strip()
        temps_domestique.append(int(ligne))

# Calcul et affichage de la moyenne
moyenne = sum(temps_domestique) / len(temps_domestique)
print("Moyenne du temps domestique (minutes) :", moyenne)
````

### 2.3 Traitement de plusieurs fichiers en lot

On peut combiner `glob` et une boucle `for` pour lire **plusieurs fichiers automatiquement**. C'est utile pour agréger des entretiens ou des journaux de terrain stockés dans des fichiers séparés.

#### Exemple — Agréger des témoignages d'entretiens


````python
import glob

# Lister tous les fichiers d'entretiens dans le dossier data/
fichiers_entretiens = glob.glob('data/entretien_*.txt')

# Dictionnaire pour regrouper tous les témoignages {Nom: Texte}
tous_les_temoignages = {}

for fichier in fichiers_entretiens:
    with open(fichier, 'r', encoding='utf-8') as f:
        for ligne in f:
            ligne = ligne.strip()
            # Les lignes commençant par '>' contiennent l'identifiant du participant
            if ligne.startswith('>'):
                nom_participant = ligne[1:]  # On enlève le '>'
                temoignage = ''
            else:
                temoignage += ligne + ' '
        tous_les_temoignages[nom_participant] = temoignage

tous_les_temoignages
````

---
### 3.3 Valeur de retour avec `return`

Une fonction peut **retourner un résultat** grâce au mot-clé `return`, ce qui permet de réutiliser la valeur dans le programme.


````python
def addition(a, b):
    return a + b

# Stocker et utiliser le résultat
total = addition(12, 15)
print(total)
````

#### `print` vs `return` — Quelle différence ?

| | `print` | `return` |
|---|---|---|
| **Rôle** | Affiche à l'écran | Envoie une valeur au programme |
| **Réutilisable ?** | ✗ Non | ✓ Oui |

**Avec `print` — la valeur n'est pas récupérable :**


````python
def saluer_print():
    print("Bonjour !")

resultat = saluer_print()
print(resultat)  # Affiche None : print() ne retourne rien
````

**Avec `return` — la valeur est récupérable :**


````python
def saluer_return():
    return "Bonjour !"

resultat = saluer_return()
resultat  # Affiche "Bonjour !"
````

### 3.4 Valeurs par défaut

On peut définir une **valeur par défaut** pour un paramètre. Elle est utilisée si aucun argument n'est fourni lors de l'appel.


````python
def saluer(prenom="participant"):
    print("Bonjour " + prenom + ", merci pour votre temps !")

saluer()           # Utilise la valeur par défaut
saluer("Thomas")   # Remplace la valeur par défaut
````

### 3.5 Arguments positionnels et par mot-clé

Quand une fonction a plusieurs paramètres, on peut passer les arguments **dans l'ordre** (positionnel) ou en **nommant les paramètres** (par mot-clé).


````python
def presentation(prenom, age):
    print("Le participant s'appelle " + prenom + ", il a " + str(age) + " ans.")

presentation("Marc", 45)       # Positionnel — ordre important
presentation(45, "Marc")       # ⚠️ Inversé : résultat incorrect
presentation(age=45, prenom="Marc")  # Par mot-clé — ordre sans importance
````

### 🖊️ Exercice 3.2 — Compter les marqueurs d'hésitation

Écrivez une fonction `compter_hesitations` qui prend un extrait d'entretien (chaîne de caractères) et retourne le nombre de marqueurs d'hésitation (`'euh'`, `'ah'`) qu'il contient. Testez-la.

> 💡 **Indice :** `.lower()` met le texte en minuscules, `.split()` divise une chaîne en liste de mots.


````python
def compter_hesitations(texte):
    marqueurs = ['euh', 'ah']
    mots = texte.lower().split(" ")
    compteur = 0
    for mot in mots:
        if mot in marqueurs:
            compteur += 1
    return compteur

# Test
extrait = "C'était euh assez difficile à euh définir."
print(compter_hesitations(extrait))
````

### 3.6 Combiner boucles et fonctions

Combiner les boucles `for` et les fonctions permet de **réduire la répétition** et d'améliorer considérablement la lisibilité du code.

#### Exemple — Anonymiser des professions socio-professionnelles

Comparez les deux approches ci-dessous :

**Sans fonction (répétitif) :**


````python
codage = {
    'Ouvrier':      'CAT_A',
    'Employé':      'CAT_B',
    'Cadre':        'CAT_C',
    'Indépendant':  'CAT_D'
}

donnees_1 = ["Ouvrier", "Cadre", "Cadre", "Employé"]
anonymise_1 = []
for profession in donnees_1:
    anonymise_1.append(codage[profession])
print(anonymise_1)

donnees_2 = ["Indépendant", "Ouvrier", "Employé", "Employé"]
anonymise_2 = []
for profession in donnees_2:
    anonymise_2.append(codage[profession])
print(anonymise_2)

donnees_3 = ["Cadre", "Indépendant", "Ouvrier", "Ouvrier"]
anonymise_3 = []
for profession in donnees_3:
    anonymise_3.append(codage[profession])
print(anonymise_3)
````

**Avec une fonction (concis et réutilisable) :**


````python
def anonymiser_categories(liste_professions):
    """Encode une liste de professions en codes de recherche anonymisés."""
    codage = {
        'Ouvrier':      'CAT_A',
        'Employé':      'CAT_B',
        'Cadre':        'CAT_C',
        'Indépendant':  'CAT_D'
    }
    codes = []
    for profession in liste_professions:
        codes.append(codage[profession])
    return codes

# Appliquer la fonction à chaque échantillon
echantillons = [
    ["Ouvrier", "Cadre", "Cadre", "Employé"],
    ["Indépendant", "Ouvrier", "Employé", "Employé"],
    ["Cadre", "Indépendant", "Ouvrier", "Ouvrier"]
]

for e in echantillons:
    print(anonymiser_categories(e))
````
